### 데이터 만들기

```python
import matplotlib.pyplot as plt
```

```python
a = [-2, 1, 2, 4, 8, 5]
b = ['A', 'B', 'C', 'D', 'E', 'F']
```

### 차트로 만들기

```python
plt.bar(b, a, color='brown')
```

```python
plt.barh(b, a, color='skyblue')
```

```python
plt.plot(b, a, 'ro-')
plt.title('subject')
plt.xlabel('school')
plt.ylabel('score')
plt.legend(['jumsu'])
```

### 한글 나타내기

```python
from matplotlib import rc
rc('font', family='malgun gothic', size=15)
```

```python
plt.plot(b, a, 'ro-')
plt.title('과목')
plt.xlabel('학교')
plt.ylabel('점수')
plt.legend(['점수'])
```
