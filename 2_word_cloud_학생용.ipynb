```
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```
# 이미지 넣기
img = '/content/drive/My Drive/prince_and_fox.png'
# 텍스트 넣기
text = open('/content/drive/My Drive/thelittleprince.txt')
```

```
Image.open(img) # 이미지 열기
```

```
lp_mask = np.array(Image.open(img))
lp_mask
```

```
# 마스킹 하기
lp_wc = WordCloud(background_color = 'white', mask = lp_mask, max_font_size = 50, contour_width = 5, contour_color = 'green').generate(text)
```

```
lp_wc.words_
```

```
# 이미지로 표현하기
lp_wc.image()
```

```
# 파일로 저장하기
lp_wc.to_file('content/drive/My Drive/1.png')
```
