### numpy를 이용한 시각화

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [1, 3, 7, 5]
# %matplotlib qt
plt.plot(y)
```

```python
x = [1, 2, 6, 7]
y = [1, 3, 7, 5]
plt.plot(x, y)
```

```python
x = [1, 2, 6, 7]
y = [x*x for x in x]
plt.plot(x, y, 'r--', marker ='o')
```

```python
a = np.arange(-10, 10, 0.1)
b = a*a
plt.plot(a, b, 'r--')
```

```python
b2 = 10 * a
plt.subplot(1, 2, 1)
plt.plot(a, b, 'r--')
plt.subplot(1, 2, 2)
plt.plot(a, b2, 'g--')
```

```python
b2 = 10 * a
plt.plot(a, b, 'r--')
plt.figure()
plt.plot(a, b2, 'g--')
```

```python
b2 = 10 * a

plt.figure(1)
plt.plot(a, b, 'r--')
plt.figure(2)
plt.plot(a, b2, 'g--')
```

```python
t  = np.arange(0, 12, 0.05)
y1 = np.sin(t)
y2 = np.cos(t)
plt.plot(t, y1)
plt.plot(t, y2)
plt.xlabel('Time')
plt.ylabel('Amp')
plt.title('SinCos')
plt.legend(['Sin', 'Cos'])
plt.grid()
```

### [문제]
- subplot을 이용해서 2행 1열 구조 2개씩
- 4가지 색상 다르게
- 4가지 스타일 다르게
- 4가지 마커 서로 다르게
- 범례 표시
- x축 y축 제목

```python
x = np.arange(0, 5, 1)
y1 = x+1
y2 = x+2
y3 = x+3
y4 = x+4

plt.subplot(1, 2, 1)
plt.plot(x, y1, 'r')
plt.plot(x, y2, 'g--')
plt.title('1st')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['Y1', 'Y2'])

plt.subplot(1, 2, 2)
plt.plot(x, y3, 'b-.')
plt.plot(x, y4, 'y-o')
plt.title('2nd')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend(['Y3', 'Y4'])
```
